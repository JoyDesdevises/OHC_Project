# Transformation Forum

```python
# Regroupement des colonnes pour moyennes + création des nouvelles colonnes

columns_activites_tribu_forum = DF_forum[['tribu1', 'tribu2', 'tribu3', 'tribu4']]
data_forum['activites_tribu'] = columns_activites_tribu_forum.mean(axis=1, skipna=True)

columns_interactions_tribu_forum = DF_forum[['tribu5', 'tribu6', 'tribu7', 'tribu8']]
data_forum['interactions_tribu'] = columns_interactions_tribu_forum.mean(axis=1, skipna=True)

data_forum['tribu_total'] = data_forum[['activites_tribu', 'interactions_tribu']].mean(axis=1, skipna=True)

columns_activites_QA_forum = DF_forum[['QA1', 'QA2', 'QA3', 'QA4']]
data_forum['activites_QA'] = columns_activites_QA_forum.mean(axis=1, skipna=True)

columns_interactions_QA_forum = DF_forum[['QA5', 'QA6', 'QA7', 'QA8']]
data_forum['interactions_QA'] = columns_interactions_QA_forum.mean(axis=1, skipna=True)

data_forum['QA_total'] = data_forum[['activites_QA', 'interactions_QA']].mean(axis=1, skipna=True)

columns_manages_forum = DF_forum[['manages1', 'manages2', 'manages3']]
data_forum['manages'] = columns_manages_forum.mean(axis=1, skipna=True)

data_forum['Es-tu manager/manageuse ?'] = DF_forum['Es-tu manager/manageuse ? ']
data_forum['Si tu es manager/manageuse, combien de managé.e.s direct(s) as-tu ?'] = DF_forum['Si tu es manager/manageuse, combien de managé.e.s direct(s) as-tu ? ']

columns_manager_forum = DF_forum[['manager1', 'manager2', 'manager3']]
data_forum['manager'] = columns_manager_forum.mean(axis=1, skipna=True)

columns_egalite_forum = DF_forum[['egalitepro1', 'egalitepro2', 'egalitepro3', 'egalitepro4']]
data_forum['egalite_pro'] = columns_egalite_forum.mean(axis=1, skipna=True)

columns_competences_forum = DF_forum[['competencesparcours1', 'competencesparcours2', 'competencesparcours3', 'competencesparcours4']]
data_forum['competences_parcours'] = columns_competences_forum.mean(axis=1, skipna=True)

columns_sante_forum = DF_forum[['santeconditions1', 'santeconditions2', 'santeconditions3', 'santeconditions4', 'santeconditions5']]
data_forum['sante_conditions_travail'] = columns_sante_forum.mean(axis=1, skipna=True)
                                                                                             
data_forum['OCTO_total'] = data_forum[['manages', 'manager', 'egalite_pro', 'competences_parcours', 'sante_conditions_travail']].mean(axis=1, skipna=True)

data_forum['statut6mois'] = DF_forum['statut6mois']

columns_missions_forum = DF_forum[['missions1', 'missions2', 'missions3', 'interco1', 'missionsinterco1', 'missionsinterco2', 'missionsinterco3', 'missionsinterco4', 'fonctionscoeur1', 'fonctionscoeur2']]
data_forum['missions_taches'] = columns_missions_forum.mean(axis=1, skipna=True)
```

```python
# Chemin relatif pour aller du dossier actuel au dossier 'Analyses par quartier'
relative_path = '../Analyses par quartier/'

# Initialiser un dictionnaire pour stocker les DataFrames
dataframes = {}

# Liste des quartiers
quartiers_mairiefrugcampus = ['Mairie', 'Frugarilla', 'Campus']

for quartier in quartiers_mairiefrugcampus:
    # Créer la clé du dictionnaire basée sur le nom du quartier
    key = f"data_{quartier}"

    # Filtrer le DataFrame pour le quartier actuel et stocker dans le dictionnaire
    dataframes[key] = data_mairie_frug_campus.loc[data_mairie_frug_campus['quartier'] == quartier]

    # Créer le chemin complet pour le fichier CSV
    filename = f"{quartier}_analyses_Novembre2023_fictif.csv"
    full_path = os.path.join(relative_path, filename)

    # Sauvegarder le DataFrame filtré dans un fichier CSV dans le dossier spécifié
    dataframes[key].to_csv(full_path, index=False)
    
# Afficher les premières lignes de chaque DataFrame stocké dans le dictionnaire
for key, dataframe in dataframes.items():
    print(f"Aperçu du DataFrame pour {key}:")
    print(dataframe.head())  # Affiche les 5 premières lignes de chaque DataFrame
    print("\n")  # Ajoute une ligne vide pour séparer les aperçus
```
